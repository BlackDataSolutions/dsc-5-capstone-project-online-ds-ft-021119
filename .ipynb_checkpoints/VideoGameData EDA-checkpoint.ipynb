## Adding in 2019 Scraped Data

- As mentioned in the intro, the dataset above was scraped from vgchartz near the end of 2016. This dataset was subsequently updated by another user with a different scraping method. 
- Below I will combine the two datasets into one cohesive dataframe with information through early 2019

```python
df_2019 = pd.read_csv('vgsales_cleaned_v3.csv')
```

```python
df_2019.head()
```

```python
df_2019.info()
```

It appears that this dataset has more missing data than the originally scraped 2016 dataset. As such, I will only bring the following years over from the 2019 set.

- 2016 (because sales data from releases in 2016 likely carried over into 2017 and 2018.)
- 2017
- 2018

After some initial exploration, I decided to drop 2019 from the dataset because there was only data for 34 games released in 2019, and most of these were either just released or haven't been released yet. 

Below, I'll first add in 2017 and 2018 because it requires no updating, and then update the 2016 sales numbers to reflect what they were in 2019.

```python
# Pulling just 2017 and 2018
updated_years_df = df_2019.loc[(df_2019['Year_of_Release'] >= 2017) & (df_2019['Year_of_Release'] < 2019)]
```

```python
updated_years_df['Year_of_Release'].value_counts()
```

```python
# Critic scores in the updated dataframe were in the format 7.8 instead of 78 (out of 100). This amends that 
updated_years_df['Critic_Score'] = updated_years_df['Critic_Score'] * 10
```

```python
# Creating an update dataframe for 2016 data
updated_sales_df = df_2019.loc[(df_2019['Year_of_Release'] == 2016)]
```

```python
len(updated_sales_df)
```

```python
updated_sales_df['2019_Global_Sales'] = updated_sales_df['Global_Sales']
```

```python
updated_sales_df['merge_name'] = updated_sales_df['Name'] + "_" + updated_sales_df['Platform']
```

```python
merger_df = updated_sales_df[['merge_name', '2019_Global_Sales']]
```

```python
#Combining the dataframes
df_list = [df_no_dupes, updated_years_df]

combined_df = pd.concat(df_list, axis=0)
```

```python
combined_df.head()
```

```python
len(combined_df)
```

```python
# Creating a merge_name column to update 2016 sales numbers
combined_df['merge_name'] = combined_df['Name'] + "_" + combined_df['Platform']
```

```python
combined_df.columns
```

```python
# Merging in updated sales data
combined_df = combined_df.merge(merger_df,how='left',on='merge_name')
```

```python
len(combined_df)
```

```python
combined_df.columns
```

```python
combined_df.count().sort_values(ascending=False)
```

```python
# Putting 0s in for all games that were not updated for 2019 numbers
combined_df['2019_Global_Sales'].fillna(0, inplace=True)
```

```python
# Replacing the global sales numbers from the 2016 set with the updated number from the 2019 set
combined_df.loc[combined_df['2019_Global_Sales'] != 0, 'Global_Sales'] = combined_df['2019_Global_Sales']
```

```python
# Checking to see if it worked
combined_df.loc[combined_df['Year_of_Release'] == 2016].sort_values('Global_Sales', ascending=False).head(20)
```

Success! I've updated all of the 2016 games with their 2019 numbers.

Still need to fill the Nans for regional sales data in with 0s. By scanning the data, I confirmed that NaNs are just placeholders for 0s. In the modeling notebooks in this project, one of the things that I try to predict is which games will be "hits." For purposes of this project, I define a "hit" as any game that sells over 1 million copies. I create this column and add it to the dataframe below.

## Genre Analysis

```python
# Previewing the data
combined_df2['Genre'].value_counts()
```

```python
# Creating genre df - global sales and hits columns
genre_df = combined_df2.groupby(['Genre'])[['Global_Sales', 'BINARY_Hits']].sum().reset_index()[['Genre', 'Global_Sales', 'BINARY_Hits']].sort_values(by='Global_Sales', ascending=False)


# Dropping Music, MMO, Board Game, Party, Visual Novel, and Sandbox since they have such small amounts
genres_to_drop = ['Music', 'Sandbox', 'Party', 'MMO', 'Visual Novel', 'Board Game']
for genre in genres_to_drop:
    genre_df = genre_df.drop(genre_df.loc[genre_df['Genre']==genre].index)
```

```python
# Also dropping in combined df2 df
for genre in genres_to_drop:
    combined_df2 = combined_df2.drop(combined_df2.loc[combined_df2['Genre']==genre].index)
```

```python
# Creating second df that counts how many games were released
genre_count = combined_df2.groupby('Genre', axis=0).count().reset_index()[['Genre','Name']].sort_values(by = "Name", ascending=True)
```

```python
genre_df = combined_df2.groupby(['Genre'])[['Global_Sales', 'BINARY_Hits']].sum().reset_index()[['Genre', 'Global_Sales', 'BINARY_Hits']].sort_values(by='Global_Sales', ascending=False)
```

```python
# Merging sales sums and count data for genre
genre_combined = pd.merge(genre_count, genre_df)
```

```python
# Creating a percent hits column that shows of the games released what percentage were hits
genre_combined['percent_hits'] = genre_combined['BINARY_Hits'] / genre_combined['Name']
```

```python
# Creating non-hits column for visualization purposes
genre_combined['non_hits'] = genre_combined['Name'] - genre_combined['BINARY_Hits']
```

```python
# Creating average sales column for visualization
genre_combined['Global_Sales_Mean'] = genre_combined['Global_Sales'] / genre_combined['Name']
```

```python
genre_combined
```

```python
def explore_df(dataframe, col_name):
    fig, ax = plt.subplots(4, 1, figsize=[20,15])
    # Sorting by total count then graphing
    dataframe = dataframe.sort_values(by='BINARY_Hits', ascending=False)
    ax[0].bar(dataframe[col_name], dataframe['non_hits'], label='Non-Hit Games')

    
    ax[0].bar(dataframe[col_name], dataframe['BINARY_Hits'], color='red', label='Hit Games')
    ax[0].set_title('Total Releases by {} (sorted by # of hits)'.format(col_name))
    ax[0].legend()
    ax[0].set_ylabel('Games Released')
    
    
    # Sorting by hit percentage then graphing
    dataframe = dataframe.sort_values(by='percent_hits', ascending=False)
    ax[1].bar(dataframe[col_name], dataframe['percent_hits'], color='darkblue')
    ax[1].set_title('Hit Percentage by {}'.format(col_name))
    ax[1].set_ylabel('Hit Percentage')
    
    # Sorting by global sales then graphing
    dataframe = dataframe.sort_values(by='Global_Sales', ascending=False)
    ax[2].bar(dataframe[col_name], dataframe['Global_Sales'], color='green')
    ax[2].set_title('Total Global Sales by {}'.format(col_name))
    ax[2].set_ylabel('Global Games Sold(millions of units)')
    
    # Sorting by average sales then graphing
    dataframe = dataframe.sort_values(by='Global_Sales_Mean', ascending=False)
    ax[3].bar(dataframe[col_name], dataframe['Global_Sales_Mean'], color='lightgreen')
    ax[3].set_title('Average Global Sales by {}'.format(col_name))
    ax[3].set_ylabel('Average Games Sold (millions of units)')
    
    plt.show
    
    
    
```

```python
explore_df(genre_combined, 'Genre')
```

## Identifying Highest Selling Platform Over Time

```python
combined_df2.columns
```

```python
# Creating platform pivot table to summarize highest selling platform by year
platform_pivot = combined_df2.pivot_table('Global_Sales', index='Platform', columns='Year_of_Release', aggfunc='sum')
# Pulling the highest performing platforms for each year
platforms = platform_pivot.idxmax()
# Bringing over their sales
annualsales = platform_pivot.max()
# Creating years list
years = platform_pivot.columns.astype(int)
# Combining data
data = pd.concat([platforms, annualsales], axis=1)
data.columns = ['Platform', 'Global_Sales']

```

```python
# Slightly modified explorer function to compensate for more categories

def explore_df_pubs(dataframe, col_name):
    fig, ax = plt.subplots(4, 1, figsize=[15,25])
    # Sorting by total count then graphing
    dataframe = dataframe.sort_values(by='BINARY_Hits', ascending=True)
    ax[0].barh(dataframe[col_name], dataframe['non_hits'], label='Non-Hit Games')

    
    ax[0].barh(dataframe[col_name], dataframe['BINARY_Hits'], color='red', label='Hit Games')
    ax[0].set_title('Total Releases by {} (sorted by # of hits)'.format(col_name))
    ax[0].legend()
    ax[0].set_xlabel('Games Released')

    # Sorting by hit percentage then graphing
    dataframe = dataframe.sort_values(by='percent_hits', ascending=True)
    ax[1].barh(dataframe[col_name], dataframe['percent_hits'], color='darkblue')
    ax[1].set_title('Hit Percentage by {}'.format(col_name))
    ax[1].set_xlabel('Hit Percentage')
    
    # Sorting by global sales then graphing
    dataframe = dataframe.sort_values(by='Global_Sales', ascending=True)
    ax[2].barh(dataframe[col_name], dataframe['Global_Sales'], color='green')
    ax[2].set_title('Total Global Sales by {}'.format(col_name))
    ax[2].set_xlabel('Global Sales Games Sold (millions of units)')
    
    # Sorting by average global sales then graphing
    dataframe = dataframe.sort_values(by='Global_Sales_Mean', ascending=True)
    ax[3].barh(dataframe[col_name], dataframe['Global_Sales_Mean'], color='lightgreen')
    ax[3].set_title('Average Global Sales by {}'.format(col_name))
    ax[3].set_xlabel('Average Global Games Sold (millions of units)')
    
    plt.show
    
    
    
```

```python
explore_df_pubs(top_20_pubs, 'Publisher')
```

## Console Wars: Sixth Generation

```python
combined_df2.columns
```

```python
def console_war_preprocessor(sony_console, nintendo_console, microsoft_console):
    gen_console_df = combined_df2.loc[(combined_df2['Platform'] == sony_console) | 
                                         (combined_df2['Platform'] == nintendo_console) |
                                         (combined_df2['Platform'] == microsoft_console)]
    
    # Counting number of games per year by console
    year_console_df = gen_console_df.groupby(['Year_of_Release','Platform'])['Name'].count().reset_index()
    
    #Pulling total revenue information (summing)
    revenue_columns = ['Global_Sales', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
    year_console_revenue = gen_console_df.groupby(['Year_of_Release','Platform'])[revenue_columns].sum().reset_index()
    
    year_combined_df = pd.merge(year_console_df, year_console_revenue)
    
    average_rating_df = gen_console_df.groupby(['Year_of_Release','Platform'])['Critic_Score'].mean().reset_index()
    
    year_combined_df2 = pd.merge(year_combined_df, average_rating_df)
    
    hit_count = gen_console_df.groupby(['Year_of_Release','Platform'])['BINARY_Hits'].sum().reset_index()
    
    year_combined_df3 = pd.merge(year_combined_df2, hit_count)
    
    return year_combined_df3
```

```python
sixth_gen_consoles = console_war_preprocessor('PS2', 'GC', 'XB')
```

```python
sixth_gen_consoles.head()
```

```python
fig = plt.figure(figsize=(12,4))

color_dict = {'PS2':'Blue', 'XB':'Green', 'GC': 'Red'}

ax1 = fig.add_subplot(1,2,1)
ax1 = sns.lineplot(x=sixth_gen_consoles['Year_of_Release'],
             y=sixth_gen_consoles['Name'],
             hue = sixth_gen_consoles['Platform'], palette=color_dict)
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Releases')

ax2 = fig.add_subplot(1,2,2)
ax2 = sns.lineplot(x=sixth_gen_consoles['Year_of_Release'],
             y=sixth_gen_consoles['BINARY_Hits'],
             hue = sixth_gen_consoles['Platform'],
                  palette=color_dict)
ax2.set_xlabel('Year')
ax2.set_ylabel('Number of Hits')

fig.subplots_adjust(hspace=0.5)
fig.suptitle('Sixth Generation Number of Releases and Hits Comparison', fontsize=15)
plt.show()
```
